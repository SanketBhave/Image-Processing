```python
import os
```

```python
files = os.listdir('cs510chips')
files
```

```python
if not os.path.exists('cs510chips/Bighorn_Sheep'):
    os.mkdir('cs510chips/Bighorn_Sheep')
if not os.path.exists('cs510chips/Bobcat'):
    os.mkdir('cs510chips/Bobcat')
if not os.path.exists('cs510chips/Coyote'):
    os.mkdir('cs510chips/Coyote')
if not os.path.exists('cs510chips/Gray_Fox'):
    os.mkdir('cs510chips/Gray_Fox')
if not os.path.exists('cs510chips/Javelina'):
    os.mkdir('cs510chips/Javelina')
if not os.path.exists('cs510chips/Mule_Deer'):
    os.mkdir('cs510chips/Mule_Deer')
if not os.path.exists('cs510chips/Raptor'):
    os.mkdir('cs510chips/Raptor')
if not os.path.exists('cs510chips/White_tailed_Deer'):
    os.mkdir('cs510chips/White_tailed_Deer')
```

```python
for file in files:
    if os.path.isdir('cs510chips/'+file):
        continue
    if 'Bighorn_Sheep' in file:
        os.rename('cs510chips/'+file, 'cs510chips/Bighorn_Sheep/'+file)
    if 'Bobcat' in file:
        os.rename('cs510chips/'+file, 'cs510chips/Bobcat/'+file)
    if 'Gray_Fox' in file:
        os.rename('cs510chips/'+file, 'cs510chips/Gray_Fox/'+file)
    if 'Javelina' in file:
        os.rename('cs510chips/'+file, 'cs510chips/Javelina/'+file)
    if 'Mule_Deer' in file:
        os.rename('cs510chips/'+file, 'cs510chips/Mule_Deer/'+file)
    if 'Raptor' in file:
        os.rename('cs510chips/'+file, 'cs510chips/Raptor/'+file)
    if 'White_tailed_Deer' in file:
        os.rename('cs510chips/'+file, 'cs510chips/White_tailed_Deer/'+file)
    if 'Coyote' in file:
        os.rename('cs510chips/'+file, 'cs510chips/Coyote/'+file)
```
