```python
import cv2
import sys
import numpy as np
import os
from os import listdir
import shutil
folder = "/s/montpelier/a/tmp/cs510bmgr8/"
animals=['Bighorn_Sheep/','Bobcat/','Gray_Fox/']

for a in animals:
    folder_dir=folder+a
    for images in os.listdir(folder_dir):
#         print(images)
        if len(images)>6:
            hour= (images.split('_')[-3])
            
            if not os.path.exists(folder_dir+'d'):
                os.mkdir(folder_dir+'d')
            if not os.path.exists(folder_dir+'n'):
                os.mkdir(folder_dir+'n')

            if (int(hour) < 4) or (int(hour) > 19):
                print(images,hour,'n')
                shutil.copy(folder_dir+images, folder_dir+'n/'+images)
            else:
                shutil.copy(folder_dir+images, folder_dir+'d/'+images)
```
